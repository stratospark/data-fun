```python
def f2(rvs=120):
    d1 = stats.poisson(2).rvs(100)
    d2 = stats.poisson(5).rvs(rvs)
    max_data = np.r_[d1, d2].max()
    bins = np.linspace(0, max_data, max_data+1)
    plt.hist(d1, bins, normed=True, color="#6495ED", alpha=.5)
    plt.hist(d2, bins, normed=True, color="#F08080", alpha=.5)
    
interact(f2, rvs=(0, 200, 10))
```

## Classifying iris class with ID3 algorithm

We wish to generate a decision tree using ID3. We will split up the 150 samples into training (50%), validation(25%), and testing data(25%).

```python
import random

#random.seed(123)
training_data_i = random.sample(range(len(data)), int(len(data)*.5))
not_training_data_i = np.setdiff1d(np.array(range(len(data))), training_data_i)
training_data = data[data.sample.isin(training_data_i)]
validation_data = data[data.sample.isin(not_training_data_i[:len(not_training_data_i)/2])]
testing_data = data[data.sample.isin(not_training_data_i[len(not_training_data_i)/2:])]

# sanity checks
print "Training Data size: " + str(len(training_data))
print "Validation Data size: " + str(len(validation_data))
print "Testing Data size: " + str(len(testing_data))
print np.intersect1d(np.intersect1d(training_data.sample, validation_data.sample), testing_data.sample)

from sklearn import tree

print "Training decision tree"
clf = tree.DecisionTreeClassifier()
clf.fit(training_data.drop("iris_class", axis=1), training_data.iris_class)

print "Validating model"
counts = (validation_data.iris_class == clf.predict(validation_data.drop("iris_class", axis=1)))
print "Accuracy: %0.3f" % (counts.value_counts().loc[True] / (1.0*len(counts)))

print "Classifying test set"
counts = (testing_data.iris_class == clf.predict(testing_data.drop("iris_class", axis=1)))
print "Accuracy: %0.3f" % (counts.value_counts().loc[True] / (1.0*len(counts)))
                                                            
# from sklearn.externals.six import StringIO
# with open("iris.dot", "w") as f:
#     f = tree.export_graphviz(clf, out_file=f)

#g.ggplot(training_data, g.aes(x="petal_length", y="petal_width", color="iris_class")) + g.geom_point()
```

```python
from sklearn.externals.six import StringIO
import pydot
dot_data = StringIO()
tree.export_graphviz(clf, feature_names=training_data.columns[1:-1], out_file=dot_data)
graph = pydot.graph_from_dot_data(dot_data.getvalue())
graph.write_png("iris.png")

from IPython.core.display import Image
Image(filename="iris.png")
```
